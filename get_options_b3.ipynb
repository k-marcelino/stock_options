#### Json direto da b3 funcionando

```python
url = 'https://www.b3.com.br/json/20221111/Posicoes/Empresa/SI_C_OPCPOSABEMP.json'
r = requests.get(url)

data = json_normalize(r.json())
data
```

```python
# Testando com uma coluna
json_normalize(data['Empresa.W'][0])
```

```python
# Concatenando todas as empresas no json
final = pd.concat([json_normalize(data[empresa][0]) for empresa in data.columns], ignore_index=True)
final
```
